```python
# # Instalacja wymaganych bibliotek
# %pip install -r ../7.\ Requirements/requirements.txt

# Instalacja nbconvert jeśli jeszcze nie zainstalowano
%pip install nbconvert

# # Uruchomienie notebooka
#!jupyter nbconvert --to notebook --execute ../1. Notebooks/01_pobieranie_i_przetwarzanie_danych.ipynb

# notebooks_to_run = [
#     "../1. Notebooks/00_okreslenie_celu_i_wymagan_projektu.ipynb",
#     "../1. Notebooks/01_pobieranie_i_przetwarzanie_danych.ipynb",
#     "../1. Notebooks/02_inicjalizacja_bazy_danych.ipynb",
#     "../1. Notebooks/03_analiza_danych.ipynb",
#     "../1. Notebooks/04_wizualizacja_dashboard_danych.ipynb",
#     "../4. ML_Models/05_pobieranie_i_przygotowanie_danych_do_modelowania.ipynb",
#     "../4. ML_Models/06_Wizualizacja_z_predykcja.ipynb"
# ]


# import os

# for notebook in notebooks_to_run:
#     print(f"Starting execution of {notebook}...")
#     command = f"jupyter nbconvert --ExecutePreprocessor.timeout=200 --to notebook --execute ../ '{notebook}'"
#     os.system(command)
#     print(f"Finished execution of {notebook}.")

import nbformat
from nbclient import NotebookClient

def run_notebook(path):
    print(f"Starting execution of {path}...")
    
    # Wczytanie notebooka
    nb = nbformat.read(path, as_version=4)
    
    # Uruchomienie notebooka
    client = NotebookClient(nb, timeout=600)  # Dostosuj timeout według potrzeb
    client.execute()
    
    # Zapisanie zmian do tego samego pliku
    nbformat.write(nb, path)
    
    print(f"Finished execution of {path}.")

# Lista ścieżek do notebooków, które chcesz uruchomić
notebooks_to_run = [
    "../1. Notebooks/00_okreslenie_celu_i_wymagan_projektu.ipynb",
    "../1. Notebooks/01_pobieranie_i_przetwarzanie_danych.ipynb",
    "../1. Notebooks/02_inicjalizacja_bazy_danych.ipynb",
    "../1. Notebooks/03_analiza_danych.ipynb",
    "../1. Notebooks/04_wizualizacja_dashboard_danych.ipynb",
    "../4. ML_Models/05_pobieranie_i_przygotowanie_danych_do_modelowania.ipynb",
    "../4. ML_Models/06_Wizualizacja_z_predykcja.ipynb"
]

for notebook_path in notebooks_to_run:
    run_notebook(notebook_path)


# # Opcjonalnie: otwórz URL
# import webbrowser
# url = 'http://localhost:8051'  # Zmień na URL generowany przez Twój notebook
# webbrowser.open(url)
```
